```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
base_ = 'https://www.edx.org'
headers_ = {"User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:109.0) Gecko/20100101 Firefox/109.0"}
```

```python
r = requests.get('https://www.edx.org/learn', headers=headers_).content
soup = BeautifulSoup(r, 'html.parser')
container_ = soup.find_all('div',class_='pgn__card-body card-body')
print(len(container_))
```

```python
def extract_URLs(container, base, headers):
    
    dixis = []
    for i in range(len(container)):
        all_as = container[i].find_all('a') 

        for a in all_as:
            url_categoria = a.get('href')
            Categoria_Curso = url_categoria[7:]

            r2 = requests.get(f'{base + url_categoria}', headers=headers).content
            soup = BeautifulSoup(r2, 'html.parser')
            container2 = soup.find_all('div',class_='discovery-card')

            if len(container2)==0:
                row = {
                    'Categoria':Categoria_Curso, 
                    'Url_Categoria':url_categoria, 
                    'Nombre_Curso':None, 
                    'URL_Curso': None}
                dixis.append(row)
            else:
                for j in range(len(container2)):
                    all_as = container2[j].find_all('a') 
                    title = container2[j].find_all('div',class_='d-card-body')
                    Nombre_Curso=title[0].find('span').contents[1].string
                    for a in all_as:
                        url = a.get('href')
                        row ={
                            'Categoria':Categoria_Curso, 
                            'Url_Categoria':base+url_categoria, 
                            'Nombre_Curso':Nombre_Curso, 
                            'URL_Curso': base + url
                        }
                        dixis.append(row)
        time.sleep(2)
    return dixis
```

```python
lista = extract_URLs(container_, base_, headers_)
```

```python
df = pd.DataFrame(columns = ['Categoria', 'Url_Categoria', 'Nombre_Curso', 'URL_Curso'])
for dixi in lista:
    df.loc[len(df)] = dixi
```
